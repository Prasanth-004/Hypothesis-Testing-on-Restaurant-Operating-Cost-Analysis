## HYPOTHESIS TESTING

## Background:
Bombay hospitality Ltd. operates a franchise model for producing exotic Norwegian dinners throughout New England. 

The operating cost for a franchise in a week (W) is given by the equation W = $1,000 + $5X, where X represents the number of units produced in a week. 

Recent feedback from restaurant owners suggests that this cost model may no longer be accurate, as their observed weekly operating costs are higher.

## Objective:
To investigate the restaurant owners' claim about the increase in weekly operating costs using hypothesis testing.


```python
# Import required libraries

import numpy as np                 # Import numpy for numerical calculations.
from scipy.stats import norm       # Import normal distribution functions.
```

## Data Provided:
●	The theoretical weekly operating cost model: W = $1,000 + $5X

●	Sample of 25 restaurants with a mean weekly cost of Rs. 3,050

●	Number of units produced in a week (X) follows a normal distribution with a mean (μ) of 600 units and a standard deviation (σ) of 25 units


```python
# Given information

sample_mean = 3050          # Sample mean weekly operating cost.

sample_size = 25            # Number of restaurants.

units_mean = 600            # Mean weekly production.

units_std = 25              # Standard deviation of production.

cost_per_unit = 5           # Cost per unit.

fixed_cost = 1000           # Fixed operating cost.

alpha = 0.05               # Significance level.
```

## Assignment Tasks:
1. State the Hypotheses statement:


```python
# State the hypotheses

print("Null Hypothesis (H0): μ = 4000")
print("Alternative Hypothesis (H1): μ > 4000")
```

First, calculate the theoretical weekly operating cost:

W=1000+5X

W=1000+5(600)
W=4000

The average weekly operating cost is equal to the theoretical cost.

Alternative Hypothesis (H₁)

Since the claim is higher than, this is a right-tailed Z-test.

## 2. Calculate the Test Statistic:
Use the following formula to calculate the test statistic(z):
where:

●	ˉxˉ = sample mean weekly cost (Rs. 3,050)

●	μ = theoretical mean weekly cost according to the cost model (W = $1,000 + $5X for X = 600 units)

●	σ = 5*25 units

●	n = sample size (25 restaurants)


```python
# Calculate theoretical mean

population_mean = fixed_cost + cost_per_unit * units_mean

print("Theoretical Mean =", population_mean)
```

```python
# Calculate population standard deviation

population_std = cost_per_unit * units_std

print("Population Standard Deviation =", population_std)
```

```python
# Calculate standard error

SE = population_std / np.sqrt(sample_size)

print("Standard Error =", SE)
```

```python
# Calculate Z-test statistic

z = (sample_mean - population_mean) / SE

print("Z Statistic =", z)
```

## 3. Determine the Probability and compare:
Using the alpha level of 5% (α = 0.05)


```python
# Calculate critical value
alpha = 0.005
critical_value = norm.ppf(1-alpha)

print("Critical Value =", critical_value)
```

For

Significance Level = 5%
Right-tailed test

Critical Value

Z
0.05 =1.645

```python
# Calculate p-value-->The p-value indicates the probability of obtaining the observed sample result if the null hypothesis is true.

p_value = 1 - norm.cdf(z)

print("P-value =", p_value)
```

## 4. Make a Decision:
Compare the test statistic with the critical value to decide whether to reject the null hypothesis.


```python
# Decision based on critical value

if z > critical_value:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

## Explanation

Decision Rule

```python
# Decision based on p-value

if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

## 5. Conclusion:

```python
# Display final conclusion
print("Conclusion")
print("--------------------------------")
print("There is no sufficient statistical evidence to support the restaurant owners' claim.")
print("Hence, we fail to reject the Null Hypothesis.")
```

## Final Conclusion

There is no statistical evidence at the 5% significance level to support the restaurant owners' claim that the weekly operating costs are higher than predicted by the theoretical model. In fact, the provided sample mean (Rs. 3050) is lower than the theoretical mean (Rs. 4000), so the data does not support the stated claim.
